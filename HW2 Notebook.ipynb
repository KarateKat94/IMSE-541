```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st
import matplotlib.pyplot as plt
```

## Question 1

```python
df = pd.read_excel("../IMSE 541 Datasets/BUSHING DIAMETER 2 BATCHES.XLS")

batch_1 = df['Batch 1']
batch_2 = df['Batch 2']
```

### Descriptive Stats and Confidence Interval

```python
print(batch_1.mean())
print(batch_1.median())
print(batch_1.std())
st.t.interval(confidence=0.975, df=len(batch_1)-1, loc=np.mean(batch_1), scale=st.sem(batch_1))
```

```python
st.norm.interval(confidence=0.975, loc=np.mean(batch_1), scale=st.sem(batch_1))
```

```python
print(batch_2.mean())
print(batch_2.median())
print(batch_2.std())
st.t.interval(confidence=0.975, df=len(batch_2)-1, loc=np.mean(batch_2), scale=st.sem(batch_2))
```

```python
st.norm.interval(confidence=0.975, loc=np.mean(batch_1), scale=st.sem(batch_1))
```
